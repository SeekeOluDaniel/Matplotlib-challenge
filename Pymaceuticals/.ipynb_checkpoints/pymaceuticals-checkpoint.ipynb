### Analysis

- The average tumor volume for mice on the Capomulin regimen is 40.68 which is the second lowest only by a 0.46 margin when compared to mice on the Ramicane regimen with a mean tumor volume of 40.22. 
- Mouse b128 responded very well to the Capomulin regimen, as the tumor volume shrunk by 6 cubic millimeters over a 35 day period and shrinking rapidly between days 25 and 35 before starting to grow and metastasize.
- There is a strong positive correlation (0.84) between mouse weight and average observed tumor volume for the entire Capomulin regimen.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_study = study_results.merge(mouse_metadata, how="outer", on="Mouse ID")


# Display the data table for preview
combined_study.head()
```

```python
# Checking the number of mice.
mice_count = len(combined_study["Mouse ID"].unique())
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = combined_study[combined_study.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_study.loc[combined_study["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_study = combined_study.drop(combined_study.loc[combined_study["Mouse ID"] == "g989"].index)
cleaned_study
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_four = cleaned_study.groupby('Mouse ID')['Timepoint'].max()
final_four = final_four.reset_index()
final_four
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
```

```python
tumor_vol = final_four.merge(cleaned_study, how="inner", on=["Mouse ID", "Timepoint"])
tumor_vol
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
    
# Create empty list to fill with tumor vol data (for plotting)
vol_data = []

for r in regimens:
    
    final_tumor_vol = tumor_vol.loc[tumor_vol["Drug Regimen"] == r, "Tumor Volume (mm3)"]
    vol_data.append(final_tumor_vol)
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{r}'s potential outliers: {outliers}")

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes

    
    # add subset 

    
    # Determine outliers using upper and lower bounds
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(avg_tumor_vol["Weight (g)"], avg_tumor_vol["Tumor Volume (mm3)"])
correlation
```

```python
print(f"The correlation between mouse weight and the average tumor volume is {correlation}")
```

```python
x_values = avg_tumor_vol["Weight (g)"]
y_values = avg_tumor_vol["Tumor Volume (mm3)"]


(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values) 

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print( line_eq )
plt.scatter(x_values,y_values)


plt.plot(x_values,regress_values,"r-")

plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")

plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
